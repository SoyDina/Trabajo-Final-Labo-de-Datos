```python
path_train = "./chest_xray/train/"
path_test = "./chest_xray/test/"
path_val = "./chest_xray/val/"
```

```python
def hist(img):
  img_to_yuv = cv2.cvtColor(img,cv2.COLOR_BGR2YUV)
  img_to_yuv[:,:,0] = cv2.equalizeHist(img_to_yuv[:,:,0])
  hist_equalization_result = cv2.cvtColor(img_to_yuv, cv2.COLOR_YUV2BGR)
  return hist_equalization_result
```

```python
path_nor_train = path_train + "NORMAL"
path_neu_train = path_train + "PNEUMONIA"

n_imagenes_nor_train = len(listdir(path_nor_train))
n_imagenes_neu_train = len(listdir(path_neu_train))
print(n_imagenes_nor_train, n_imagenes_neu_train)
#n_imagenes_neu_train = n_imagenes_nor_train #Hago esto para tener el mismo n√∫mero de datos de 
                                            #una clase que de la otra

imagenes_normal = np.zeros(shape=(n_imagenes_nor_train, 224, 224))
imagenes_pneumonia = np.zeros(shape=(n_imagenes_neu_train, 224, 224)) 

y_nor = np.zeros(shape=(n_imagenes_nor_train))
y_neu = np.ones(shape=(n_imagenes_neu_train))

for condicion in ["NORMAL", "PNEUMONIA"]:
    path = path_train + condicion
    lista_directorios = listdir(path)
    print(condicion, len(lista_directorios))
    
    for index, im in enumerate(lista_directorios):
        imagen = cv2.imread(path + "/" + im)

        imagen = hist(imagen) #Sube el contraste de las imagenes
        imagen = cv2.resize(imagen, (224,224))[:,:,0]
        
        if condicion == "NORMAL":
            imagenes_normal[index] += imagen
        elif condicion == "PNEUMONIA":
            imagenes_pneumonia[index] += imagen
        
        clear_output(wait=True)
        print(str(100*(index+1)/(len(lista_directorios))) + "%")
        print(index, path)



print(imagenes_normal[0].shape, len(imagenes_pneumonia), type(imagenes_normal[0]))
```

```python
def augment(images_copiar, y_images_copiar, n_rot, categoria=0):
    images = np.copy(images_copiar)
    y_images = np.copy(y_images_copiar)
    
    for ii, image in enumerate(images):
        img_flip = np.array([cv2.flip(image, 1)])
        images = np.concatenate((images, img_flip))
        y_images = np.append(y_images, categoria)
        
        clear_output(wait=True)
        print(ii, "de", images.shape[0])

    for i in range(n_rot):
        img = np.copy(images[i])
        img_rot = ndimage.rotate(img, 5)
        img_rot = np.array([cv2.resize(img_rot, (224,224))])
        images = np.concatenate((images, img_rot))
        y_images = np.append(y_images, categoria)
        clear_output(wait=True)
        print(i, "de", n_rot)
    return images, y_images
```
